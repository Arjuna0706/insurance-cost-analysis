# **Import Lib**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, train_test_split
```

# **Data Load**

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("harishkumardatalab/medical-insurance-price-prediction")

print("Path to dataset files:", path)

csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
file_path = os.path.join(path, csv_file)

df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.info()
```

```python
df.replace('?', np.nan, inplace=True)

df['sex'] = df['sex'].map({'female': 0, 'male': 1})

df['smoker'] = df['smoker'].map({'no': 0, 'yes': 1})

df['region'] = df['region'].map({'southwest': 0, 'southeast': 1, 'northwest': 2, 'northeast': 3})

df['bmi'] = df['bmi'].astype(float)

df.head(10)
```

```python
df.info()
```

# **Exploratory Data Analysis**

```python
sns.regplot(x='bmi', y='charges', data=df, line_kws={"color": "red"})
plt.ylim(0,)
```

Through the regplot visualization, a positive correlation can be observed between Body Mass Index (BMI) and insurance charges. The upward-sloping trend line indicates that higher BMI values are generally associated with increased medical and insurance costs. However, the relatively wide dispersion of data points suggests that BMI is not the sole determinant of insurance charges, implying interactions with other factors, particularly smoking status.

```python
sns.boxplot(x='smoker', y='charges', data=df, hue='smoker', palette='Set2', legend=False)
```

The boxplot clearly illustrates a substantial difference in the distribution of insurance charges between smokers and non-smokers. The median insurance cost for the smoker group is significantly higher than that of the non-smoker group. Interestingly, the lower boundary of the smoker group's box is nearly aligned with the upper outlier range of the non-smoker group. This visual evidence strongly confirms that smoking status is the most influential risk factor contributing to a significant increase in insurance costs.

# **Model Development**

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
X = df[['smoker']]
Y = df[['charges']]
lm.fit(X,Y);
```

```python
lm.score(X,Y)
```

```python
Z = df[['smoker', 'bmi', 'sex', 'age', 'children', 'region']]
lm.fit(Z,Y);
```

```python
lm.score(Z,Y)
```

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model', LinearRegression())]
pipe=Pipeline(Input)
Z = Z.astype(float)
pipe.fit(Z,Y)
ypipe=pipe.predict(Z)
print(r2_score(Y,ypipe))
```

# **Model Evaluation and Refinement**

```python
x_train, x_test, y_train, y_test = train_test_split(Z, Y, test_size=0.2, random_state=1)

ridge=Ridge(alpha=0.1)
ridge.fit(x_train,y_train)
ridge.score(x_test,y_test)
```

```python
pr = PolynomialFeatures(degree=2)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.transform(x_test)
ridge.fit(x_train_pr, y_train)
y_hat = ridge.predict(x_test_pr)
print(r2_score(y_test,y_hat))
```

Conclusion
1. Performance Summary
Based on the experimental results, the simple linear regression model and the initial Ridge Regression model achieved moderate R-squared scores of approximately 0.677. However, the model's performance improved significantly after data transformation. By applying Polynomial Features (degree = 2), the model was able to capture non-linear relationships among variables, resulting in an increase of the R-squared score to 0.786 on the test dataset.

2. Best Model Selection
The most optimal model selected in this project is the Pipeline Model, which integrates StandardScaler, PolynomialFeatures (degree = 2), and LinearRegression. This model achieved the highest performance with an R-squared score of 0.846 on the training dataset. The results indicate that interactions between features such as the relationship between BMI and smoking status play a significant role in improving the accuracy of insurance cost predictions.

3. Practical Insights
With these evaluation results, the model has proven to be sufficiently reliable for automatically estimating patients' insurance charges based on demographic characteristics and health-related factors. The implementation of this model can assist healthcare providers and insurance companies in risk management and in determining insurance premiums more accurately through a data-driven approach.
